```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
grafico = plt.figure()
ax = grafico.add_subplot(1, 1, 1)
ax.plot(np.random.rand(10))
plt.show()
```

```python
grafico_2 = plt.figure(figsize=(12, 6))
plt.show()
```

```python
grafico_2 = plt.figure(figsize=(12, 6))
plt.title('Meu Grfico Personalizado', fontsize=12, fontweight='bold')
plt.xlabel('Eixo X', fontsize=10)
plt.ylabel('Eixo Y', fontsize=10)
plt.plot(np.random.rand(20), color='blue', marker='o', linestyle='--', linewidth=2, markersize=6, label='Dados Aleatrios 1')
plt.plot(np.random.rand(20), color='red', marker='o', linestyle='--', linewidth=2, markersize=6, label='Dados Aleatrios 2')
plt.grid(True)
plt.legend()
plt.show()
```

```python
grafico.savefig('grafico.png')
```

```python
value = np.random.rand(5)
labels = ['A', 'B', 'C', 'D', 'E']

plt.pie(value, labels=labels, autopct='%1.1f%%', startangle=90, shadow=True, explode=(0, 0, 0.1, 0, 0))
plt.title('Grfico de Pizza Exemplo')
plt.show()
```
