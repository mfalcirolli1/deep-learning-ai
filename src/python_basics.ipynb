## Variáveis

Variáveis são usadas para armazenar valores. Em Python, você não precisa declarar o tipo da variável explicitamente.

```python
x = 10
print(x)
```

## Tipos de Dados

Python suporta vários tipos de dados básicos, como inteiros, floats, strings e booleanos.

```python
inteiro = 5
flutuante = 3.14
texto = "Olá"
booleano = True
print(type(inteiro), type(flutuante), type(texto), type(booleano))
```

## Operadores

Operadores aritméticos incluem adição (+), subtração (-), multiplicação (*), divisão (/), e divisão inteira (//).

```python
a = 10
b = 3
print(a + b, a - b, a * b, a / b, a // b)
```

## Estruturas de Controle

Use `if` e `else` para tomar decisões baseadas em condições.

```python
if a > b:
    print("a é maior")
else:
    print("b é maior ou igual")
```

## Loops

O loop `for` é usado para iterar sobre uma sequência.

```python
for i in range(5):
    print(i)
```

## Funções

Funções são blocos de código reutilizáveis definidos com `def`.

```python
def saudacao(nome):
    return f"Olá, {nome}!"

print(saudacao("Mundo"))
```

## Listas

Listas são estruturas de dados mutáveis que podem armazenar vários itens.

```python
lista = [1, 2, 3, 4]
print(lista)
lista.append(5)
print(lista)
```

## Classes

Classes são blueprints para criar objetos. Elas permitem definir atributos e métodos.

```python
class Pessoa:
    def __init__(self, nome, idade):
        self.nome = nome
        self.idade = idade
    
    def saudacao(self):
        return f"Olá, meu nome é {self.nome} e tenho {self.idade} anos."

pessoa1 = Pessoa("João", 25)
print(pessoa1.saudacao())
```
